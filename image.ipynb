# Stable Diffusion XL Image Generation
This notebook uses the `diffusers` library to generate an image from a text prompt using **Stable Diffusion XL**.

```python
# Install dependencies (uncomment if running for the first time)
# !pip install diffusers transformers accelerate safetensors torch --upgrade
```

```python
from diffusers import DiffusionPipeline
import torch
```

## Load the Pretrained Model

```python
pipe = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,
    use_safetensors=True,
    variant="fp16"
)
pipe.to("cuda")
```

### Optional: Enable Memory-Efficient Attention
Use this if you are on PyTorch < 2.0 for reduced GPU memory usage.

```python
# pipe.enable_xformers_memory_efficient_attention()
```

## Generate an Image from Prompt

```python
prompt = "An astronaut riding a green horse"
images = pipe(prompt=prompt).images[0]

# Display the generated image
images.show()
```
