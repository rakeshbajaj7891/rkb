# Multi-Agent Quiz Generator using Groq (LLaMA 8B)

This notebook builds a multi-agent system that generates quizzes based on Board, Class, and Subject.

```python
from groq import Groq

# 🔑 Replace with your API key
client = Groq(api_key="YOUR_GROQ_API_KEY")
```

## LLM Call Function

```python
def call_llm(prompt):
    response = client.chat.completions.create(
        model="llama3-8b-8192",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7
    )
    return response.choices[0].message.content
```

## Multi-Agent Components

```python
# 🧠 Input Analyzer Agent
def input_analyzer(user_input):
    prompt = f"""
    Extract structured JSON from: {user_input}
    Include: board, class, subject, number of questions
    """
    return call_llm(prompt)

# 📚 Curriculum Agent
def curriculum_agent(data):
    prompt = f"""
    Based on this data: {data}
    List important syllabus topics.
    """
    return call_llm(prompt)

# ✍️ Question Generator Agent
def question_generator(data, topics):
    prompt = f"""
    Generate a quiz based on:
    {data}
    Topics: {topics}
    Include MCQs and short answers with answers.
    """
    return call_llm(prompt)

# ✅ Validator Agent
def validator(quiz):
    prompt = f"""
    Validate this quiz for correctness and clarity:
    {quiz}
    Fix any mistakes.
    """
    return call_llm(prompt)
```

## Orchestrator

```python
def run_system(user_input):
    print("🔍 Analyzing input...")
    data = input_analyzer(user_input)

    print("📚 Fetching curriculum...")
    topics = curriculum_agent(data)

    print("✍️ Generating questions...")
    quiz = question_generator(data, topics)

    print("✅ Validating quiz...")
    final_quiz = validator(quiz)

    return final_quiz
```

## Run Example

```python
user_input = "CBSE Class 8 Science 5 questions"

result = run_system(user_input)
print(result)
```
